```python
#Problem 1
a = int(input(''))
b = int(input(''))
c = int(input(''))
aTwo = a * a
bTwo = b * b
cTwo = c * c
if aTwo + bTwo > cTwo:
    print(str(a) + ' squared plus ' + str(b) + ' squared is greater than ' + str(c) + ' squared')
if aTwo + bTwo < cTwo:
    print(str(a) + ' squared plus ' + str(b) + ' squared is less than ' + str(c) + ' squared')
if aTwo + bTwo == cTwo:
    print(str(a) + ' squared plus ' + str(b) + ' squared is equal to ' + str(c) + ' squared')
```
